## Boxplot
Boxcox helps to find how the data is distributed based on fetal health status to find any relationship. It also identifies outliers to assess data trasformation. 

### Boxplot on features based on fetal health status 


```python
x_value = data['fetal_health']
for i in data.select_dtypes(np.number).columns:
    plt.figure(figsize = (9, 6))
    sns.boxplot(x = x_value, y = i, data = data, order = ['A','B','C'], palette = 'rocket')
    plt.title("Fetal health VS")
    
    
```

Findings:

+ accelarations:
    + A has acceleration majority between 0.0 - 0.0070
    + B and C rarely has accelaration but some outliers are found
    + Above findings indicates that accelaration may have a high correlation to A

    
+ uterine_contractions:
    + uterine contraction below 0.002 falls under fetal health B or C
    
    
+ prolonged_decelerations:
    + Most data points between approx. 0.000 - 0.002 fall under C
    + Above findings indicates that prolonged deceleration may has high correlation to C

+ histogram_width:
    + The width for A is between approx.45 - 100 with evenly distributed 

+ histogram_variance:
    + High variance for C compared to A and B

+ histogram_tendency:
    + tendency with the '-' values are categorises as C

## lmplot 
Here I apply lmplot (linear model plot) to visualize the relation of two variables. 
First I will plot histogram_width vs histogram number of peaks , histogram min where we found high correlation from the correlation metrix. Later I will plot the several features with prolongued decelaration to find patterns for each fetal health status.

###  Histogram width vs histogram number of peaks , histogram min

```python
columns = ['histogram_number_of_peaks', 'histogram_min' ]

for i in columns:
    
    g = sns.lmplot(x= 'histogram_width', y = i,
           hue = 'fetal_health', data = data,
           palette = 'rocket',
           hue_order = ['A', 'B', 'C'],
           markers = ['o', 'd', 'x'],
           col = 'fetal_health',
           col_order = ['A', 'B', 'C'],
           x_bins = 20,
           #scatter_kws = {'s': 150, 'edgecolor':'b'}
          )
    g.set_xlabels('histogram_width')
    g.set_ylabels(i)
```

Findings:
+ histogram_number_of_peaks:
    + Data points in A are less distributed than B and C
    + The width increases together with the number of peaks

### Prolongued decelerations vs 
+ accelerations
+ histogram width
+ histogram_max
+ histogram_number_of_peaks
+ histogram_variance


```python
columns = ['accelerations', 'histogram_width','histogram_max','histogram_number_of_peaks', 'histogram_variance']
for i in columns:
    g = sns.lmplot(x= i, y = 'prolongued_decelerations',
               hue = 'fetal_health', data = data,
               palette = 'rocket',
               hue_order = ['A', 'B', 'C'],
               markers = ['o', 'd', 'x'],
               col = 'fetal_health',
               col_order = ['A', 'B', 'C'],
               x_bins = 20,
               #scatter_kws = {'s': 150, 'edgecolor':'b'}
              )
    g.set_xlabels(i)
    g.set_ylabels('prolongued_decelerations')
```


Findings:
+ A shows in all graph a flat line
+ in contrast,prolongued decelerations increases in B and C
+ In C the distribution of the data points are wide

## Relplot
+ Relplot is usuful tool to visualize the relationship of 2 variables.
+ The line shows the mean of the data points and connects into a line

*Please note that the columns below are the ones which shows great results to see the relationship to the y xis.
Relationship was checked with all columns but due to the running time of the query, the plots with good results found are listed.*

### Baseline values vs
+ histogram_mean
+ histogram__variance


```python
columns = ['histogram_mean', 'histogram_variance']

for i in columns:
        
    g = sns.relplot(x = 'baseline_value', y= i,
             data = data,
                    col = 'fetal_health',
                    col_order = ['A', 'B', 'C'],
                palette = 'rocket',
                    
             hue = 'fetal_health',
                   hue_order = ['A', 'B', 'C'])
    g.set_xlabels('Fetal heart beats per minute')
    g.set_ylabels(i)
```


Findings:
+ histogram mean
    + In the upper graph (baseline vs histogram mean), the histogram mean increases when the baseline value increases
    + The histogram mean of A is rather higher than B and C
    + Low histogram mean can be an indicater of C
+ histogram variance
    + The histogram variance data points are widely spread in C

### Severe decelerations vs other numerical features

```python
columns = [
       'uterine_contractions', 'light_decelerations', 
       'prolongued_decelerations', 'abnormal_short_term_variability',
       'mean_value_of_short_term_variability',
       'histogram_width', 'histogram_number_of_peaks',
       'histogram_number_of_zeroes','histogram_variance']

for i in columns:
        
    g = sns.relplot(x = i, y= 'severe_decelerations',
             data = data,
                    col = 'fetal_health',
                    col_order = ['A', 'B', 'C'],
                kind = 'line',
                    palette = 'rocket',
             hue = 'fetal_health',
                   hue_order = ['A', 'B', 'C'])
    g.set_xlabels(i)
    g.set_ylabels('Severe decelerations')
```

Findings:
   + The larger the light deceleration it becomes, it increases the severe deceleration in C
   + The smaller the prolonged deceleration is, the larger of the severe decelerations in A and particularly in C
   + The smaller the histogram number of zeros the more severe decelerations for C
   + More occurence of severe decelerations in histogram mean between 70 -90 in C
   + Above findings indicates that severe decelerations have high correlation to multiple features which can help detect fetal health C
